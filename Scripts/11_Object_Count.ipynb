```python
import os
from glob import glob

# === CONFIGURATION ===
DATASET_ROOT = r"\Dataset_Repository"  # change this to your main folder

total_images = 0
total_objects = 0
class_object_count = {}

# Loop through each class folder
for class_folder in sorted(os.listdir(DATASET_ROOT)):
    class_path = os.path.join(DATASET_ROOT, class_folder)
    if not os.path.isdir(class_path):
        continue

    txt_files = glob(os.path.join(class_path, "*.txt"))
    img_files = glob(os.path.join(class_path, "*.jpg")) + glob(os.path.join(class_path, "*.png"))
    
    class_total_objects = 0
    for txt in txt_files:
        with open(txt, "r") as f:
            lines = [line.strip() for line in f if line.strip()]
            class_total_objects += len(lines)
            total_objects += len(lines)
    
    class_object_count[class_folder] = class_total_objects
    total_images += len(img_files)

# Compute average
avg_objects_per_image = total_objects / total_images if total_images > 0 else 0

# === OUTPUT ===
print(f"Total Images: {total_images}")
print(f"Total Objects: {total_objects}")
print(f"Average Objects per Image: {avg_objects_per_image:.2f}\n")

print("Objects per Class:")
for cls, count in class_object_count.items():
    print(f"  {cls}: {count}")
```
