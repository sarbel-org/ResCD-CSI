```python
import os

# === CONFIG ===
base_dataset_path = r"Dataset_Repository"  # top-level folder containing class folders
image_extensions = [".jpg", ".jpeg"]

# === LOOP THROUGH SUBFOLDERS ===
for folder_name in os.listdir(base_dataset_path):
    folder_path = os.path.join(base_dataset_path, folder_name)
    if not os.path.isdir(folder_path):
        continue

    print(f"Processing folder: {folder_name}")

    # Get all images and txt files
    files = os.listdir(folder_path)
    images = [f for f in files if os.path.splitext(f)[1].lower() in image_extensions]
    txts = [f for f in files if f.endswith(".txt")]

    # Create dicts for matching
    image_basenames = {os.path.splitext(f)[0]: f for f in images}
    txt_basenames = {os.path.splitext(f)[0]: f for f in txts}

    counter = 1  # numbering per folder

    # Rename files with matching image + txt
    matched = set(image_basenames.keys()) & set(txt_basenames.keys())
    for base in matched:
        new_name_base = f"{folder_name}_FI_{counter}"
        # Rename image
        old_img = os.path.join(folder_path, image_basenames[base])
        new_img = os.path.join(folder_path, new_name_base + os.path.splitext(image_basenames[base])[1].lower())
        while os.path.exists(new_img):
            counter += 1
            new_name_base = f"{folder_name}_FI_{counter}"
            new_img = os.path.join(folder_path, new_name_base + os.path.splitext(image_basenames[base])[1].lower())
        os.rename(old_img, new_img)

        # Rename txt
        old_txt = os.path.join(folder_path, txt_basenames[base])
        new_txt = os.path.join(folder_path, new_name_base + ".txt")
        while os.path.exists(new_txt):
            counter += 1
            new_name_base = f"{folder_name}_FI_{counter}"
            new_txt = os.path.join(folder_path, new_name_base + ".txt")
        os.rename(old_txt, new_txt)

        counter += 1

    # Images without txt → rename to delete_FI_<number>
    unmatched_images = set(image_basenames.keys()) - matched
    for base in unmatched_images:
        new_name_base = f"delete_FI_{counter}"
        old_img = os.path.join(folder_path, image_basenames[base])
        new_img = os.path.join(folder_path, new_name_base + os.path.splitext(image_basenames[base])[1].lower())
        while os.path.exists(new_img):
            counter += 1
            new_name_base = f"delete_FI_{counter}"
            new_img = os.path.join(folder_path, new_name_base + os.path.splitext(image_basenames[base])[1].lower())
        os.rename(old_img, new_img)
        counter += 1

    # Txt without images → rename to delete_FI_<number>
    unmatched_txts = set(txt_basenames.keys()) - matched
    for base in unmatched_txts:
        new_name_base = f"delete_FI_{counter}"
        old_txt = os.path.join(folder_path, txt_basenames[base])
        new_txt = os.path.join(folder_path, new_name_base + ".txt")
        while os.path.exists(new_txt):
            counter += 1
            new_name_base = f"delete_FI_{counter}"
            new_txt = os.path.join(folder_path, new_name_base + ".txt")
        os.rename(old_txt, new_txt)
        counter += 1

print("All files renamed successfully!")
```
